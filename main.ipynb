## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
# your code here
for w in words:
  print(w.upper())
```

**Create a new list containing only words with 5 or more letters**

```python
# your code here
words_with_5_or_more_letters = []
for w in words:
  if len(w) >= 5:
    words_with_5_or_more_letters.append(w)
```

**Print the first word starting with "t"**

```python
# your code here
for w in words:
  if w[0] == 't':
    print(w)
    break
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
# your code here
squares = []
for i in range(1, 11):
  squares.append(i**2)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
# your code here
squares_of_odd_numbers = []
for i in range(1, 11):
  if (i%2 != 0):
    squares_of_odd_numbers.append(i**2)

print(squares_of_odd_numbers)
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
# your code here
squares_of_multiples_of_8 = []
for num in range(8,1001, 8):
  squares_of_multiples_of_8.append(num**2)

print(squares_of_multiples_of_8)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
# your code here
len(people)
```

**How many people have kids**?

```python
# your code here
count = 0
for person in people:
  if person['n_kids'] > 0:
    count += 1

print(count)
```

**How many kids do they have in total?**

```python
# your code here
count = 0
for person in people:
  count += person['n_kids']

print(count)
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
# your code here
people_next_year = []
for person in people:
  new_person = person.copy()
  if new_person['name'][-1] == 'a':
    new_person['n_kids'] += 1
  people_next_year.append(new_person)

print(people_next_year)


```
